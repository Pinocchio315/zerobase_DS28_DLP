```python
from PIL import Image
import os

# 이미지 크기를 320x320으로 조정하는 함수
def resize_to_320(image_path, output_path):
    # 이미지 열기
    img = Image.open(image_path)
    
    # 정사각형으로 만들기 위해 가장 긴 쪽을 기준으로 이미지 확장
    max_size = max(img.size)
    
    # 정사각형 배경 생성 (흰색)
    square_img = Image.new('RGB', (max_size, max_size), (255, 255, 255))
    
    # 이미지 중앙에 배치
    square_img.paste(img, ((max_size - img.width) // 2, (max_size - img.height) // 2))
    
    # 320x320으로 리사이즈
    resized_img = square_img.resize((320, 320))
    
    # 변환된 이미지 저장
    resized_img.save(output_path)
```

- 폴더에 있는 이미지 사이즈 한번에 변환

```python
# 이미지를 폴더별로 저장해놓은 전체 폴더 경로
input_folder = './'

# 이미지 저장 폴더 이름
want_folder = ['갈치', '고등어', '닭고기', '새우', '소고기', '연어', '오징어', '조개류']

# 원하는 이미지 폴더만 작업
for folder in want_folder:

    output_folder = f'{input_folder}resize_{folder}'
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    folder_path = input_folder + folder
    
    for idx, file_name in enumerate(os.listdir(folder_path)):

        # 파일 경로 설정
        input_path = os.path.join(folder_path, file_name)

        if folder == '갈치':
            new_name = 'hairtail'
        elif folder == '고등어':
            new_name = 'mackerel'
        elif folder == '닭고기':
            new_name = 'chicken'
        elif folder == '새우':
            new_name = 'shrimp'
        elif folder == '소고기':
            new_name = 'beef'
        elif folder == '연어':
            new_name = 'salmon'
        elif folder == '오징어':
            new_name = 'squid'
        elif folder == '조개류':
            new_name = 'shellfish'

        output_path = os.path.join(output_folder, new_name + f'_{idx+1}.jpg')
        print(output_path)
        
        # 이미지 리사이즈 및 저장
        resize_to_320(input_path, output_path)
        #print(f"Processed {filename}")
```

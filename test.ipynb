## 리스트

```python
model = YOLO('./YOLODataset/best.pt')  

# 이미지 폴더 경로 
image_folder_path = './YOLODataset/images/test'  
image_files = glob.glob(os.path.join(image_folder_path, '*.*'))  

# 리스트
ingredients = []

# 이미지에 대해 예측
for image_path in image_files:
    # 모델 예측
    results = model.predict(image_path, imgsz=320, conf=0.3)

    # 예측된 결과 출력
    print(f"Results for {image_path}:")
    print(results)  

    # 예측된 객체들의 클래스 이름을 추출하여 리스트에 추가
    for result in results:  
        print("Prediction details:")
        print(f"Boxes: {result.boxes}")  
        print(f"Names: {result.names}")  

        # 바운딩 박스에서 클래스 인덱스와 관련된 정보를 추출
        for i, box in enumerate(result.boxes.xywh):  
            class_idx = int(result.boxes.cls[i].cpu().numpy())  
            ingredient_name = result.names[class_idx]  
            ingredients.append(ingredient_name)

# 결과 출력 (중복 제거)
ingredients = list(set(ingredients))  
print("Detected ingredients:", ingredients)
```
